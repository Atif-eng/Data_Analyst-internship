> <h3>Check for Missing Values</h3>

```python
missing_value=data.isnull().sum().sum()
if missing_value>0:
    display(Markdown('<h2 style="color: red;"> Missing Values Found</h2>'))
    print(data.isnull().sum())
else:
    display(Markdown('<h2 style="color: green;"> Missing Values Not Found </h2>'))
```

> <h3>Impute Missing Values</h3>

```python
data["CouponCode"].unique()
```

```python
data["CouponCode"]=data["CouponCode"].fillna("No_Coupon")
```

```python
missing_value=data.isnull().sum().sum()
if missing_value>0:
    display(Markdown('<h2 style="color: red;"> Missing Values </h2>'))
    print(data.isnull().sum())
else:
    display(Markdown('<h2 style="color: green;">Missing Values Removed</h2>'))
    print(data.isnull().sum())

display(Markdown('<h3 style="color: green;"> Imputation Completed! </h3>'))
display(Markdown(f'<h4 style="color: yellow;"> Remaining null values in <b>CouponCode</b>: {data["CouponCode"].isnull().sum()}</h4>'))
```

> <h3>Handling Extra Spaces</h3>

```python
display(Markdown('<h2 style="color: purple;">Remove extra Spaces done!</h2>'))
categorical_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']
for col in categorical_cols:
    data[col] = data[col].astype(str).str.strip()
```

> <h3>Save Cleaned Dataset</h3>

```python
clean_data=data.to_csv("clean_data.csv")
```
