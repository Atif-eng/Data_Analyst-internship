```python
sns.boxplot(x=clean_data['TotalPrice'], color='orangered')
plt.title('Outlier in Total Order Price')
plt.xlabel('Total Price')
plt.show()

# Calculate IQR to identify exact outlier thresholds
Q1 = clean_data['TotalPrice'].quantile(0.25)
Q3 = clean_data['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + (1.5 * IQR)

outliers = clean_data[clean_data['TotalPrice'] > upper_bound]
print(f"Number of outlier orders (> ${upper_bound:.2f}): {len(outliers)}")
```

```python
outliers = clean_data[clean_data['TotalPrice'] > upper_bound]
normal_data = clean_data[clean_data['TotalPrice'] <= upper_bound]

plt.scatter(
    normal_data.index, 
    normal_data['TotalPrice'], 
    color='steelblue', 
    alpha=0.5, 
    s=25, 
    label='Normal Orders'
)

plt.scatter(
    outliers.index, 
    outliers['TotalPrice'], 
    color='red', 
    marker='x', 
    s=70, 
    linewidths=2, 
    label=f'Outliers (> ${upper_bound:.2f})'
)

plt.axhline(
    y=upper_bound, 
    color='orangered', 
    linestyle='--', 
    linewidth=1.5, 
    label=f'IQR Threshold Boundary (${upper_bound:.2f})'
)

# 5. Polish the aesthetics
plt.title('Identified Outliers', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Order Sequence', fontsize=11)
plt.ylabel('Total Transaction Value ($)', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper right', frameon=True, shadow=True)

# Render
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales by Month-Year to see macroeconomic trends
clean_data["Date"] = pd.to_datetime(clean_data["Date"], errors="coerce")
clean_data["YearMonth"] = clean_data["Date"].dt.to_period("M")
monthly_sales = clean_data.groupby('YearMonth')['TotalPrice'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)

plt.figure(figsize=(14, 5))
sns.lineplot(data=monthly_sales, x='YearMonth', y='TotalPrice', marker='o', color='purple')
plt.xticks(rotation=45)
plt.title('Revenue Trend Analysis (2023 - 2025)',fontsize=20)
plt.ylabel('Total Sales Revenue ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# --- 1. Calculate Revenue Drivers ---
product_sales = clean_data.groupby('Product')['TotalPrice'].sum()
top_product = product_sales.idxmax()
top_product_rev = product_sales.max()
bottom_product = product_sales.idxmin()
bottom_product_rev = product_sales.min()

# --- 2. Calculate Customer Behavior (Mean vs Median) ---
mean_val = clean_data['TotalPrice'].mean()
median_val = clean_data['TotalPrice'].median()

# Determine skewness / outlier direction
if mean_val > median_val:
    skew_explanation = "high-value outliers (large individual purchases) are pulling the average upward"
else:
    skew_explanation = "low-value bulk orders are pulling the average downward"

# --- 3. Identify Trend Performance ---
# Convert date to Year-Month period for grouping
clean_data['YearMonth'] = pd.to_datetime(clean_data['Date']).dt.to_period('M')
monthly_sales = clean_data.groupby('YearMonth')['TotalPrice'].sum()
peak_month = monthly_sales.idxmax()
peak_month_val = monthly_sales.max()

# Format the peak month (e.g., "January 2024")
peak_month_str = peak_month.to_timestamp().strftime('%B %Y')

# --- 4. Operational Risk Metrics ---
status_counts = clean_data['OrderStatus'].value_counts()
returned_count = status_counts.get('Returned', 0)
cancelled_count = status_counts.get('Cancelled', 0)
total_orders = len(clean_data)
risk_percentage = ((returned_count + cancelled_count) / total_orders) * 100

# Find high-risk referral channels for returns/cancellations
risk_orders = clean_data[clean_data['OrderStatus'].isin(['Returned', 'Cancelled'])]
referral_risk = risk_orders['ReferralSource'].value_counts(normalize=True) * 100
top_risk_source = referral_risk.idxmax() if not referral_risk.empty else "Unknown"
top_risk_pct = referral_risk.max() if not referral_risk.empty else 0

# --- 5. Generate Completed Markdown ---
completed_summary = f"""
## 📊 Executive Summary & Key Observations
Based on the descriptive analysis of the {total_orders:,} orders in this dataset:

* **Revenue Drivers:** **{top_product}** represents the highest volume of sales (generating ${top_product_rev:,.2f}), whereas **{bottom_product}** is lagging behind (generating ${bottom_product_rev:,.2f}).
* **Customer Behavior:** The average transaction value (Mean) is **${mean_val:.2f}**, but the Median is **${median_val:.2f}**, indicating that **{skew_explanation}**.
* **Trend Performance:** Revenue experienced a significant growth trend peaking in **{peak_month_str}** (reaching ${peak_month_val:,.2f} in monthly sales), likely driven by high transaction counts and the utilization of the `SAVE10` or `FREESHIP` coupon codes. 
* **Operational Risk:** We noticed that **{risk_percentage:.1f}%** of total orders ({returned_count} returned, {cancelled_count} cancelled) did not result in a successful completion. Our referral source analysis indicates that traffic coming from **{top_risk_source}** represents the highest share of these lost conversions (**{top_risk_pct:.1f}%** of all cancelled/returned orders), which warrants immediate operational review.
"""

display(Markdown(completed_summary))
```
