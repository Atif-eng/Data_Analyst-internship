> <h3>Check for Missing Values</h3>

```python
missing_value=data.isnull().sum().sum()
if missing_value>0:
    display(Markdown('<h2 style="color: red;"> Missing Values Found</h2>'))
    print(data.isnull().sum())
else:
    display(Markdown('<h2 style="color: green;"> Missing Values Not Found </h2>'))
```

> <h3>Impute Missing Values</h3>

```python
data["CouponCode"].unique()
```

```python
data["CouponCode"]=data["CouponCode"].fillna("No_Coupon")
```

```python
missing_value=data.isnull().sum().sum()
if missing_value>0:
    display(Markdown('<h2 style="color: red;"> Missing Values </h2>'))
    print(data.isnull().sum())
else:
    display(Markdown('<h2 style="color: green;">Missing Values Removed</h2>'))
    print(data.isnull().sum())

display(Markdown('<h3 style="color: green;"> Imputation Completed! </h3>'))
display(Markdown(f'<h4 style="color: yellow;"> Remaining null values in <b>CouponCode</b>: {data["CouponCode"].isnull().sum()}</h4>'))
```

> <h3>Handling Extra Spaces</h3>

```python
display(Markdown('<h2 style="color: purple;">Remove extra Spaces done!</h2>'))
categorical_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']
for col in categorical_cols:
    data[col] = data[col].astype(str).str.strip()
```

> <h3>Set Date</h3>

```python
data["Date"]=pd.to_datetime(data["Date"])
display(Markdown('<h2 style="color: purple;">Date set done!</h2>'))
```

> <h3>Save Cleaned Dataset</h3>

```python
clean_data=data.to_csv("clean_data.csv")
```

```python
numerical_cols = clean_data.select_dtypes(include=['number']).columns
for col in numerical_cols:
    sns.histplot(clean_data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    sns.boxplot(x=clean_data[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
categorical_cols = clean_data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if clean_data[col].nunique() < 50:
        sns.countplot(y=clean_data[col],
                       order = clean_data[col].\
                        value_counts().\
                            index)
        
        plt.title(f'Count Plot of {col}')
        plt.xlabel('Count')
        plt.ylabel(col)
        plt.show()
```

```python
correlation_matrix = clean_data[numerical_cols].corr()
sns.heatmap(correlation_matrix,
             annot=True,
             cmap='coolwarm', 
             fmt=".2f")

plt.title('Correlation Matrix')
plt.show()
```
